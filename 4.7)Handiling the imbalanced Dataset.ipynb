```python
 #Import the dependencies
 import pandas as pd
 import numpy as np
```

```python
#Loading the data set in to the padas data frame
a = pd.read_csv("/content/credit_data.csv")
```

```python
a.head()
```

```python
a.tail()
```

```python
#distribution of the two classes
a['Class'].value_counts()
```

**This is high imbalanced dataset**

*   0--> Legit Transactions
*   1--> Fraudulent Transactions



```python
#separating the legit transactions and fraudulent transactions
legit = a[a.Class == 0]
fraudulent = a[a.Class == 1]
```

```python
print(legit.shape)
print(fraudulent.shape)
```

**Under Sampling**

*   Build a sample dataset containing similar distributions of legit and fraudulent Transactions. 
* Number of fraudulent Transactions --> 492




```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

**Concatenate the Two DataFrames**

```python
b = pd.concat([legit_sample, fraudulent], axis = 0)
```
