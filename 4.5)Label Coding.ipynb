**Label Coding**

*   converting the label into numeric forms



```python
#Importing the Dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

**Label Encoding of breast cancer data set**

```python
#Loading the data from csv file to pandas dataframe
c= pd.read_csv("/content/data.csv")
```

```python
#First five rows of the data frame
c.head()
```

```python
#Finding the count of different labels
c['diagnosis'].value_counts()
```

```python
#load the lable encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(c.diagnosis)
```

```python
#Appending the labels to the DataFrame
c['target'] = labels
```

```python
c.head()
```

**0 -->Bengin**

**1 -->Malignant**

```python
c['target'].value_counts()
```
